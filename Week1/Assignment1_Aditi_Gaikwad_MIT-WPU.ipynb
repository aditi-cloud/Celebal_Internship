##### 3. Handle missing values (identify, fill/drop).

```python
print(df.isnull().sum())
```

```python
# Fill text/object columns

df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna('No customer reviews')

df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

df['videos'] = df['videos'].fillna('No Videos')

df['seller_information'] = df['seller_information'].fillna('Seller information not available')

df['variations'] = df['variations'].fillna('No Variations')
```

##### 4. Perform basic operations (filter rows, select columns). 

```python
#practicing filtering of rows
df_cheap_products = df[df['initial_price'] < 1000]
print(df_cheap_products)
```

```python
df_filtered=df[['product_id', 'title','initial_price', 'discount']]
# Calculating price after discount,[assuming discount is in absolute terms and not percentage]
df_filtered['price'] = df_filtered['initial_price'] - df_filtered['discount']
df_filtered.head()
```

```python
#since quantity column is missing, i am creating a new column quantity and filling it with random values between 1 and 10 for each row
import numpy as np
df_filtered['quantity'] = np.random.randint(1, 11, size=len(df_filtered))
df_filtered.head()
```

##### 5. Remove duplicates.

```python
df_filtered.drop_duplicates(subset='product_id', inplace=True)
print(len(df_filtered))
```

##### 6. Create a derived column (total_amount = price * quantity). 

```python
df_filtered['total_amount'] = df_filtered['price'] * df_filtered['quantity']
df_filtered.head()
```

```python
total_amount_sum = df_filtered['total_amount'].sum()
print("Total amount sum:", total_amount_sum)
```
