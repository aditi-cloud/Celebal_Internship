##### 3. Handle missing values (identify, fill/drop).

```python
df['final_price'] = df['final_price'].str.replace(r'[^\d.]', '', regex=True)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce')

print(df[['initial_price', 'discount', 'final_price']].dtypes)
```

```python
print(df.isnull().sum())
```

```python

df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna('No customer reviews')

df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

df['videos'] = df['videos'].fillna('No Videos')

df['seller_information'] = df['seller_information'].fillna('Seller information not available')

df['variations'] = df['variations'].fillna('No Variations')
```

##### 4a. Feature Engineering

```python
import numpy as np

df['price_difference'] = df['initial_price'] - df['final_price']
df['popularity_metric'] = df['rating'] * np.log1p(df['ratings_count'])
df[['title', 'initial_price', 'final_price', 'price_difference', 'popularity_metric']].head()
```

Univariate Analysis: Distribution of Ratings

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Bivariate Analysis: Initial Price vs. Final Price

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(data=df, x='initial_price', y='final_price', hue='rating', palette='viridis', alpha=0.7)
plt.title('Bivariate Analysis: Initial Price vs. Final Price')
plt.xlabel('Initial Price (INR)')
plt.ylabel('Final Price (INR)')
plt.show()
```

Category-level Analysis: Top 10 Popular Categories

```python
top_categories = df.groupby('category')['popularity_metric'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='crest')
plt.title('Top 10 Most Popular Product Categories')
plt.xlabel('Average Popularity Metric')
plt.ylabel('Category')
plt.show()
```

##### 4B. Filtering rows and columns

```python
#practicing filtering of rows
df_cheap_products = df[df['initial_price'] < 1000]
print(df_cheap_products.head())
```

```python
df_filtered=df[['product_id', 'title','final_price', 'discount']]
df_filtered.rename(columns={'final_price': 'price'}, inplace=True)
df_filtered.head()
```

```python
#since quantity column is missing, i am creating a new column quantity and filling it with random values between 1 and 10 for each row
import numpy as np
df_filtered['quantity'] = np.random.randint(1, 11, size=len(df_filtered))
df_filtered.head()
```

##### 5. Remove duplicates.

```python
df_filtered.drop_duplicates(subset='product_id', inplace=True)
print(len(df_filtered))
```

##### 6. Create a derived column (total_amount = price * quantity). 

```python
df_filtered['total_amount'] = df_filtered['price'] * df_filtered['quantity']
df_filtered.head()
```

```python
total_amount_sum = df_filtered['total_amount'].sum()
print("Total amount sum:", total_amount_sum)
```
